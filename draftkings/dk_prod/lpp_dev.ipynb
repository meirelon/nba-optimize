```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://www.draftkings.com/lineup/getavailableplayerscsv?contestTypeId=70&draftGroupId=22401")[["ID", "Name", "Position", "TeamAbbrev", "Salary", "AvgPointsPerGame"]]
```

```python
data_col_names = ["Id", "Name", "Position", "Team", "Salary", "AvgPtsPerGame"]
df.columns = data_col_names
df = df.set_index("Name")
```

```python
query = pd.read_pickle("../query.pkl")
model = pd.read_pickle("model.pkl")
```

```python
prepared_query = query.format(season=2019, dt='2018-10-27', partition_date='20181028')
today_df = pd.read_gbq(prepared_query, project_id='scarlet-labs', dialect="standard", verbose=False).fillna(value=0)
today_df = today_df.set_index("player")
```

```python
today_df.shape
```

```python
prediction_input = today_df.select_dtypes([np.number]).drop(['dk', 'secs_played'], axis=1).dropna()
today_df["Projected"] = model.predict(prediction_input)
```

```python
tmp = df.join(today_df["Projected"], how="left").reset_index().drop(["AvgPtsPerGame"], axis=1).fillna(value=0)
```
